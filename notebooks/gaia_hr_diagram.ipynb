## Imports

```python
import astroquery
from astroquery.gaia import Gaia
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## ADQL query

```python
query = """ 
SELECT TOP 30000
    source_id, 
    ra, 
    dec, 
    parallax, 
    parallax_error,
    phot_g_mean_mag, 
    phot_bp_mean_mag,
    phot_rp_mean_mag,
    bp_rp,
    ruwe,
    teff_gspphot
    FROM gaiadr3.gaia_source
    WHERE parallax > 5
        AND parallax_over_error > 10
        AND phot_g_mean_mag IS NOT NULL
        AND phot_bp_mean_mag IS NOT NULL
        AND phot_rp_mean_mag IS NOT NULL
        AND bp_rp IS NOT NULL
    """
```

## Send the query to Gaia

```python
job = Gaia.launch_job_async(query)
gaia_table = job.get_results()
```

## Convert and inspect

```python
df = gaia_table.to_pandas()
df.head()
```

## Check the dataset

```python
print(df.shape)
df.info()
```

## Save data

```python
df.to_csv("../data/raw/gaia_nearby_stars_raw.csv", index=False)
```
